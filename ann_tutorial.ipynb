```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
```

```python
class MLP(nn.Module):
    def __init__(self,n):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(n,n),
            nn.ReLU(),
            nn.Linear(n,n),
            nn.ReLU(),
            nn.Linear(n,n)
        )
    
    def forward(self,x):
        return self.layers(x)

    
    
class prep:
    def __init__(self,n):
        self.n = n
        print(f'number of dimentions: {n}')
        x = np.linspace(0,2*np.pi,n)
        self.x = x
    
    def gen(self,num,delta):
        self.num = num
        self.delta = delta
        print(f'generating {num} pairs of data')
        
        y_i = np.zeros([num,self.n])
        y_o = np.zeros([num,self.n])
        
        for i in range(num):
            y_input = np.double(np.sin(self.x - 2*np.pi*np.random.rand()))
            y_output = np.double(np.sin(self.x - 2*np.pi*np.random.rand() - delta))
            y_i[i,:] = y_input
            y_o[i,:] = y_output
        
        self.i = y_i
        self.o = y_o
```

```python
p = prep(101)
p.gen(100,0.5)
np.double(0)
type(p.i[0][1])
```

```python
n = 101
num_epoch = 100
if __name__ == '__main__':
    torch.manual_seed(42)
    
    mlp = MLP(n)
    loss_function = nn.MSELoss()
    optimizer = torch.optim.Adam(mlp.parameters(),lr=1e-4)
    
    for epoch in range(0,num_epoch):
        
        print(f'epoch: {epoch+1}')
        
        for n,i in enumerate(p.i):
            o = p.o[n]
            
            i_t = torch.from_numpy(i).float()
#             display(i_t)
            
            o_t = torch.from_numpy(o).float()
            o_model = mlp(i_t)
            loss = loss_function(o_model,o_t)
            print(loss)
            loss.backward()
            optimizer.step()
```
